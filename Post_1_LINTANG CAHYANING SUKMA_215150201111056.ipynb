# Modul

```python
def binstep(x, th=0):
  return 1 if x >= th else 0
```

```python
import numpy as np

def MCP(x, w, th):
  y_in = np.dot(x, w)
  y_out = binstep(y_in, th)
  return y_out
```

```python
def calc_accuracy(a, b):
  s = [1 if a[i] == b[i] else 0 for i in range(len(a))]
  return sum(s) / len(a)
```

```python
def AND(X):
  w = 1, 1
  th = 2
  y = [MCP(i, w, th) for i in X]
  return y
```

```python
data = (0, 0), (0, 1), (1, 0), (1, 1)
output = AND(data)
true = 0, 0, 0, 1
accuracy = calc_accuracy(output, true)

print('Output:', output)
print('True:', true)
print('Accuracy:', accuracy)
```

```python
def OR(X):
  w = 2, 2
  th = 2
  y = [MCP(i, w, th) for i in X]
  return y
```

```python
data = (0, 0), (0, 1), (1, 0), (1, 1)
output = OR(data)
true = 0, 1, 1, 1
accuracy = calc_accuracy(output, true)

print('Output:', output)
print('True:', true)
print('Accuracy:', accuracy)
```

```python
def ANDNOT(X):
  w = 2, -1
  th = 2
  y = [MCP(i, w, th) for i in X]
  return y
```

```python
data = (0, 0), (0, 1), (1, 0), (1, 1)
output = ANDNOT(data)
true = 0, 0, 1, 0
accuracy = calc_accuracy(output, true)

print('Output:', output)
print('True:', true)
print('Accuracy:', accuracy)
```

```python
def XOR(X):
  X_flip = [(i[1], i[0]) for i in X]
  y1 = ANDNOT(X)
  y2 = ANDNOT(X_flip)
  y = zip(y1, y2)
  z = OR(y)
  return z
```

```python
data = (0, 0), (0, 1), (1, 0), (1, 1)
output = XOR(data)
true = 0, 1, 1, 0
accuracy = calc_accuracy(output, true)

print('Output:', output)
print('True:', true)
print('Accuracy:', accuracy)
```

Kode di bawah ini merupakan pengembangan dari JST McCulloch-Pitts Neuron. Disini teman-teman akan menyelesaikan masalah logika yang merupakan kombinasi dari dari logika-logika yang telah dibuat fungsi sebelumnya. Selesaikan dengan memanggil fungsi-fungsi tersebut.

```python
#template soal
def logic(X):
    f0 = [x[0] for x in X]
    f1 = [x[1] for x in X]
    f2 = [x[2] for x in X]

    op1 = [f0[i] and not (f1[i] and f2[i]) for i in range(len(X))]
    op2 = [f0[i] or (f1[i] and f2[i]) for i in range(len(X))]
    op3 = [(f0[i] and not f1[i]) or (not f0[i] and f2[i]) for i in range(len(X))]
    op4 = [f1[i] and f1[i] and f1[i] for i in range(len(X))]

    # Kombinasi operasi logika
    final_output = zip(op4, op1)

    return op2

data = ((0, 0, 0), (0, 1, 0), (0, 0, 1), (0, 1, 1), (1, 1, 1), (1, 0, 1), (1, 0, 0), (1, 1, 0))
true = (0, 0, 1, 1, 1, 1, 1, 0)
output = logic(data)
accuracy = calc_accuracy(output, true)

print(accuracy)
```

f0, f1, dan f2 adalah tiga input logika biner. Pada soal ini mencoba menggabungkan logika-logika sebelumnya untuk mengahsilkan nilai akurasi yang sudah ditentukan. Variabel op1, op2, op3, dan op4 adalah hasil dari operasi-operasi logika yang digunakan untuk mengkombinasikan input ini untuk menghasilkan keluaran akhir.
