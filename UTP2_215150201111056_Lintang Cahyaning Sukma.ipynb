## MADALINE

```python
import numpy as np
```

```python
def aktivasi(x):
  if x < 0:
    return -1
  else:
    return 1
```

```python
def test(w,v,b,test_data):
  v_aktivasi = np.vectorize(aktivasi)
  z_in = np.dot(test_data, w)
  z_in = z_in + b[:-1]
  z = v_aktivasi(z_in)
  y_in = np.dot(z, v) + b[-1]
  y = v_aktivasi(y_in)
  return y
```

```python
def calc_accuracy(a, b):
  s = [1 if a[i] == b[i] else 0 for i in range(len(a))]
  return sum(s) / len(a)
```

Tugas praktikan adalah melengkapi kode Madaline dibawah sehingga dapat dijalankan dengan jumlah hidden neuron yang bervariasi. Praktikan dapat menggunakan kode Madaline yang ada pada modul sebagai referensi.

```python
def train(train_data,train_target,alpha=0.1,max_epoch=10,n=2):
  #n adalah jumlah hidden neuron yang ingin dipakai
  #inisialisasi bobot hidden layer (yang menghubungkan 2 input neuron dengan n hidden neuron)
  w = np.random.rand(train_data.shape[1], n)

  #bobot ouput layer bernilai 0.5 sejumlah hidden neuron
  v = np.full(n, 0.5)

  #inisialisasi bias berjumlah n, bernilai 0.5
  b = np.full(n, 0.5)

  #menghubungkan dengan bias output neuron
  b = np.append(b,0.5)
  epoch = 0
  v_aktivasi = np.vectorize(aktivasi)
  weight_updated = True
  while weight_updated == True and epoch < max_epoch:
    weight_updated = False
    for data,target in zip(train_data,train_target):
      z_in = np.dot(data,w)
      z_in = z_in + b[:-1]
      z = v_aktivasi(z_in)
      y_in = np.dot(z,v) + b[-1]
      y = v_aktivasi(y_in)
      if y!= target:
        weight_updated = True
        if target == 1:
          #Perubahan bobot pada hidden neuron dengan input mendekati 0
          w += alpha * data[:, np.newaxis]
          b[:-1] += alpha

        elif target == -1:
          index = np.where(z_in>0)[0]
          #Perubahan bobot pada setiap hidden neuron yang memiliki input positif (praktikan bisa melakukan iterasi untuk semua b dan w dengan indeks pada variabel index)
          w[:, index] -= alpha * data[:, np.newaxis]
          b[index] -= alpha

    epoch = epoch +1
  return (w,v,b)
```

Pengujian

```python
data = np.array([[-1,-1],[-1,1],[1,-1],[1,1]])
target = np.array([-1,1,1,1])
(w,v,b) = train(data,target,alpha=0.2,max_epoch=10, n=2)
output = test(w,v,b,data)
accuracy = calc_accuracy(output, target)

print('Output:', output)
print('Target:', target)
print('Accuracy:', accuracy)
```
